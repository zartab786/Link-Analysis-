```python
#reading the data 
import math
fp=open(r"dataset1.txt","r")
list_lines=[]
while True:
    line=fp.readline()
    if(not line):
        break
    list_lines.append(line)
print(len(list_lines))
```

```python
#creating the edge list 
edge_list=[]
for line in list_lines:
    text=line.split()
    text[0]=int(text[0])
    text[1]=int(text[1])
    edge_list.append(text)
print("Total Number of edges :" ,len(edge_list))
```

```python
#first counting the total nodes
dict_nodes_in={}
dict_nodes_out={}
dict_nodes_total={}
for edge in edge_list:
    u=edge[0]
    v=edge[1]
    
    if u in dict_nodes_out:
        dict_nodes_out[u]+=1
    else:
        dict_nodes_out[u]=1
    
    if v in dict_nodes_in:
        dict_nodes_in[v]+=1
    else:
        dict_nodes_in[v]=1
    
    if u not in dict_nodes_total:
        dict_nodes_total[u]=1
    
    if v not in dict_nodes_total:
        dict_nodes_total[v]=1

print("Total Nodes are : ",len(dict_nodes_total))
# print(len(dict_nodes_in))
# print(len(dict_nodes_out))
```

```python
#calculating degrees average, maximum etc.
avg_in_degree=0
avg_out_degree=0
for node in dict_nodes_in:
    avg_in_degree+=dict_nodes_in[node]

avg_in_degree=avg_in_degree/len(dict_nodes_total)
print("Average in degree is :",avg_in_degree)

for node in dict_nodes_out:
    avg_out_degree+=dict_nodes_out[node]

avg_out_degree=avg_out_degree/len(dict_nodes_total)
print("Average out degree is :",avg_out_degree)
```

```python
import operator
sorted_dict_in=dict((sorted(dict_nodes_in.items(), key =operator.itemgetter(1), reverse=True)))
sorted_dict_out=dict((sorted(dict_nodes_out.items(), key =operator.itemgetter(1), reverse=True)))
for i in sorted_dict_in:
    print("Max indegree node is : ",i)
    break
for i in sorted_dict_out:
    print("Max outdegree node is : ",i)
    break
```

```python
#calculating density of graph
density=len(edge_list)/((len(dict_nodes_total))*(len(dict_nodes_total)-1))
print("The density of newtork is:",density)
                        
```

```python
#plotting distribution of degrees
from matplotlib import pyplot as plt
list_of_nodes=[i for i in range(0,len(dict_nodes_total))]
list_indegree=[]
for i in range(0,len(dict_nodes_total)):
    if i in dict_nodes_in:
        list_indegree.append(dict_nodes_in[i])
    else:
        list_indegree.append(0)
list_outdegree=[]
for i in range(0,len(dict_nodes_total)):
    if i in dict_nodes_out:
        list_outdegree.append(dict_nodes_out[i])
    else:
        list_outdegree.append(0)
plt.title("In degree distribution")
plt.plot(list_of_nodes,list_indegree)
plt.xlabel("Nodes in Graph")
plt.ylabel("Degree Values")
plt.show()

plt.title("Out degree distribution")
plt.plot(list_of_nodes,list_outdegree)
plt.xlabel("Nodes in Graph")
plt.ylabel("Degree Values")
plt.show()
```

```python
# creating adjacency matrix
adj_matrix=[[0 for i in range(0,len(dict_nodes_total))]for j in range(0,len(dict_nodes_total))]
print(len(adj_matrix[0]))
```

```python
for edge in edge_list:
    u=edge[0]
    v=edge[1]
    adj_matrix[u][v]=1

print(len(adj_matrix[1]))
```

```python
#calculating the clustering coefficient of each node
dict_clust_cof={}
for i in range(0,len(dict_nodes_total)):
    adj_lis=adj_matrix[i]
    count_neighb=0
    count_neighb_edg=0
    for node in range(0,len(adj_lis)):
        if(adj_matrix[i][node]==1):
            count_neighb+=1
            for node2 in range(0,len(adj_lis)):
                if(node!=node2 and adj_matrix[node][node2]==1 and (adj_matrix[i][node2]==1 or adj_matrix[node2][i]==1)):
                    count_neighb_edg+=1
        elif(adj_matrix[node][i]==1):
            count_neighb+=1
            for node2 in range(0,len(adj_lis)):
                if(node!=node2 and adj_matrix[node][node2]==1 and (adj_matrix[i][node2]==1 or adj_matrix[node2][i]==1)):
                    count_neighb_edg+=1
        
    
    count_neighb=(count_neighb*(count_neighb-1))
    if(count_neighb_edg!=0 and count_neighb!=0):
        hold=count_neighb_edg/count_neighb
        
        dict_clust_cof[i]=hold
    else:
        dict_clust_cof[i]=0

print("The clustering coefficient of nodes :")
print((dict_clust_cof))

# for val in dict_clust_cof:
#     if(dict_clust_cof[val]==1):
#         count+=1
# print(count)       
                    
   
            
```

```python
coeff_list=[dict_clust_cof[i] for i in dict_clust_cof ]
plt.title("Clustering Coefficient Distribution")
plt.plot(list_of_nodes,coeff_list)
plt.xlabel("Nodes in Graph")
plt.ylabel("Coefficient Values ")
plt.show()
```

```python
import pickle
#reading the distance matrix
fp=open("dist_matrix","rb")
dist_matrix=pickle.load(fp)
fp.close()
```

```python
#taking average of all the shortest paths
#if for any node centrality is 25572 here it means some nodes are unreachable to from that node
dict_centrality={}
for i in range(0,len(dict_nodes_total)):
    sum_dist=0
    count_reach=0
    for j in range(0,len(dict_centrality)):
        
        #nodes that can reach i 
        if(i!=j and dist_matrix[j][i]!=25572):
            count_reach+=1
            sum_dist+=dist_matrix[j][i]
            
    
    if(sum_dist!=0):
        sum_dist=sum_dist/(count_reach)
    
    dict_centrality[i]=sum_dist
            
    
print("The closeness centrality of the nodes : ")
print(dict_centrality)
```

```python
#distribution of closeness centrality
cent_list=[dict_centrality[i] for i in dict_centrality ]
plt.title("Closeness Centrality Distribution")
plt.plot(list_of_nodes,cent_list)
plt.xlabel("Nodes in Graph")
plt.ylabel("Centrality Values ")
plt.show()
```

```python
import numpy as np
import networkx as nx
#converting into numpy array
adj2=np.array(adj_matrix)
print(adj2)
```

```python
#creating graph of matrix
G=nx.from_numpy_matrix(adj2)
#print(G.edges(data=True))
```

```python
#calculating pagerank score for each node
page_score=nx.pagerank(G,max_iter=300)
```
